```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
player_demos = purchase_data.loc[:, ["Gender", "SN", "Age"]]
player_demos = player_demos.drop_duplicates()
player_numbers = player_demos.count()[0]
player_numbers
pd.DataFrame({"Total Numbers": [player_numbers]})
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
print("Total Purchases ", purchase_data["Purchase ID"].count())
```

```python
print("Total Revenue ", purchase_data["Price"].sum())
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_counts = player_demos["Gender"].value_counts()
```

```python
gender_percentage = gender_counts/player_numbers*100
gender_demographics_percent = pd.DataFrame({"Total Count": gender_counts, "Percent Players": gender_percentage})
gender_demographics_percent
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_purchase_total = purchase_data.groupby(["Gender"]).sum()["Price"].rename("Total Purchase Amount")
gender_purchase_avg = purchase_data.groupby(["Gender"]).mean()["Price"].rename("Avg Purchase Amount")
gender_purchase_count = purchase_data.groupby(["Gender"]).count()["Price"].rename("Count Purchases")
```

```python
normalized_totals = gender_purchase_total/ gender_demographics_percent["Total Count"]
normalized_totals
```

```python
gender_data = pd.DataFrame({"Purchase Count": gender_purchase_total, "Purchase Average": gender_purchase_avg, "Purchase Count": gender_purchase_count, "Average per Person": normalized_totals})
gender_data
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
player_demos["Age Range"] = pd.cut(player_demos["Age"], age_bins, labels = group_names)
player_demos.head()
```

```python
age_demo_total = player_demos["Age Range"].value_counts()
age_demo_avg = age_demo_total/player_numbers*100

age_demo = pd.DataFrame({"Age Demo Total": age_demo_total, "Average Age": age_demo_avg})
age_demo.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data["Age Range"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)

age_purchase_total = purchase_data.groupby(["Age Range"]).sum()["Price"].rename("Total Purchase Amount")
age_purchase_avg = purchase_data.groupby(["Age Range"]).mean()["Price"].rename("Avg Purchase Amount")
age_purchase_count = purchase_data.groupby(["Age Range"]).count()["Price"].rename("Count Purchases")



```

```python
normalized_totals = age_purchase_total/ age_demo["Age Demo Total"]
normalized_totals
```

```python
gender_data = pd.DataFrame({"Age Purchase Count": age_purchase_total, "Purchase Avg by Age": age_purchase_avg, "Purchase Count": age_purchase_count, "Average per Person": normalized_totals})
gender_data
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order

```python
user_total = purchase_data.groupby(["SN"]).sum()["Price"].rename("Total Purchase Amount")
user_avg = purchase_data.groupby(["SN"]).mean()["Price"].rename("Avg Purchase Amount")
user_count = purchase_data.groupby(["SN"]).count()["Price"].rename("Count Purchases")
user_data = pd.DataFrame({"Purchase Count": user_total, "Avg Purchse Price": user_avg, "Total Purchases": user_count})
user_data.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
most_popular = purchase_data.loc[:, ["Item ID", "Item Name", "Price"]]
most_popular
```

```python
mp_purchase_total = most_popular.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Purchase Amount")
mp_purchase_avg = most_popular.groupby(["Item ID", "Item Name"]).mean()["Price"].rename("Avg Purchase Amount")
mp_purchase_count = most_popular.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Count Purchases")

mp_data_pd = pd.DataFrame({"Total Purchase Amount": mp_purchase_total, "Avg Purchase Amount": mp_purchase_avg, "Count Purchases": mp_purchase_count})
mp_data_pd_sorted = mp_data_pd.sort_values("Count Purchases", ascending=False)
mp_data_pd_sorted.head()

```
